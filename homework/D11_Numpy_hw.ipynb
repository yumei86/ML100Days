將以下問卷資料的職業(Profession)欄位缺失值填入字串'others'，更進一步將字串做編碼。

```python
import pandas as pd
q_df = pd.DataFrame([['male', 'teacher'], 
              ['male', 'engineer'],
              ['female', None],
              ['female', 'engineer']],columns=['Sex','Profession'])
q_df
```

```python
#缺失值填入字串'others'
new_q_df = q_df.fillna('others')

new_q_df
```

```python
#更進一步將字串做編碼。 此時用什麼方式做編碼比較適合?為什麼?
#ans: One-hot Encoding(一位有效編碼)，性別及職業類別無順序性，因此用此法分類較佳。
pf = pd.get_dummies(new_q_df[['Sex']])
new_q_df = pd.concat([new_q_df, pf], axis=1)
new_q_df
```
