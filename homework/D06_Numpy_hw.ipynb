多陣列存一起需要存成npz，讀取須注意npz中有多個陣列

題目:<br>
1. 將下兩列array存成npz檔<br>
array1 = np.array(range(30))<br>
array2 = np.array([2,3,5])<br>
2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔


```python
#. 將下兩列array存成npz檔
import numpy as np

array1 = np.array(range(30))
array2 = np.array([2,3,5])

# wb : 以二進制寫入檔案 | rb : 以二進制讀取檔案

with open('hw6.npy', 'wb') as f:
    np.save(f, array1)
    np.save(f, array2)
```

```python
#2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔

#自己設定的
array3 = np.array([7,8,9,10])

with open('hw6.npy', 'rb') as f:
    a = np.load(f)
    b = np.load(f)

with open('new_hw6.npy', 'wb') as p:
    np.savez(p,a,b,array3)
```
