```python
import numpy as np 
```

```python
#1.正常的談話的聲壓為20000微巴斯卡，請問多少分貝?
#請寫下程式
V1 = 20000
V0 = 20
ans = 20 * np.log10((V1/V0))
print('ans =',ans,'GdB')
```

```python
#2.30分貝的聲壓會是50分貝的幾倍?
#公式移項過後可以得到 V1 = ?
#請寫下程式
#V1 = ((GdB/20)**10) * V0
V0 = 20
V30 = ((30/20)**10) * V0
V50 = ((50/20)**10) * V0
ans = V30/V50
print('ans =',ans)
```
