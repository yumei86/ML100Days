作業目標<br>
計算有缺失值的資料，統計量實作<br>
作業重點<br>
當遇到缺失值有函式可以處理，不須額外寫程式刪除<br>
計算統計量時不能出現缺失值

題目:<br>
english_score = np.array([55,89,76,65,48,70])<br>
math_score = np.array([60,85,60,68,np.nan,60])<br>
chinese_score = np.array([65,90,82,72,66,77])<br>
上3列共六位同學的英文、數學、國文成績，第一個元素代表第一位同學，舉例第一位同學英文55分、數學60分、國文65分，今天第五位同學因某原因沒來考試，導致數學成績缺值，運用上列數據回答下列問題。<br>
1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
3. 用補考後資料找出與國文成績相關係數最高的學科?

```python
import numpy as np
```

```python
english_score = np.array([55,89,76,65,48,70])
math_score = np.array([60,85,60,68,np.nan,60])
chinese_score = np.array([65,90,82,72,66,77])
```

```python
#1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
lis = [english_score, math_score, chinese_score]
i = 0

for item in lis:
    if not np.isnan(np.sum(item)):
        name = ['英文','數學','國文']
        
        print(name[i],"陣列中無 nan，忽略 nan 後的平均值為：", np.mean(item))
    else:
        print(name[i],"陣列中有 nan，忽略 nan 後的平均值為：", np.nanmean(item))
    i +=1

i = 0
for item in lis:
    if not np.isnan(np.sum(item)):
        name = ['英文','數學','國文']
        
        print(name[i],"陣列中無 nan，忽略 nan 後的最大值為：", np.max(item))
    else:
        print(name[i],"陣列中有 nan，忽略 nan 後的最大值為：", np.nanmax(item))
    i +=1

i = 0
for item in lis:
    if not np.isnan(np.sum(item)):
        name = ['英文','數學','國文']
        
        print(name[i],"陣列中無 nan，忽略 nan 後的最小值為：", np.min(item))
    else:
        print(name[i],"陣列中有 nan，忽略 nan 後的最小值為：", np.nanmin(item))
    i +=1

i = 0
for item in lis:
    if not np.isnan(np.sum(item)):
        name = ['英文','數學','國文']
        
        print(name[i],"陣列中無 nan，忽略 nan 後的標準差為：", np.std(item))
    else:
        print(name[i],"陣列中有 nan，忽略 nan 後的標準差為：", np.nanstd(item))
    i +=1
```

```python
#2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
math_score[4] = 55
print('補考後的平均值 =',np.mean(math_score))
print('補考後的最大值 =',np.max(math_score))
print('補考後的最小值 =',np.min(math_score))
print('補考後的標準差 =',np.std(math_score))
```

```python
#3. 用補考後資料找出與國文成績相關係數最高的學科?
a = np.corrcoef(chinese_score,math_score)[0,1]
b = np.corrcoef(chinese_score,english_score)[0,1]
if a>b:
    print('ans:數學')
else:
    print('ans:英文')
```
